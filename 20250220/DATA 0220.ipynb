```python
from astropy.io import fits
import matplotlib.pyplot as plt
from astropy.time import Time
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
import glob

folder1 = "/content/drive/MyDrive/Colab Notebooks/RT/20260220/FITS"

fits_files1 = sorted(glob.glob(f"{folder1}/*.fits"))
```

```python
for fname in fits_files1:
    with fits.open(fname) as hdul:
        data = hdul[1].data
        hdr = hdul[0].header  # Access header from primary HDU (index 0)
        x1 = data["Az_Offset"]
        x2 = data["El_Offset"]
        y = data["LEFT_POL"]
        y1=data["RIGHT_POL"]
        #t = Time(data["JD"][0], format="jd")
        #time_str = t.iso.split(".")[0]
        az = np.median(data["Azimuth"])
        el = np.median(data["Elevation"])

        plt.figure(figsize=(12,6))
        plt.plot(x1, y, ".", ms=4)
        plt.title(fname.split("/")[-1])
        plt.xlabel("Az_Offset")
        plt.ylabel("LEFT_POL")
        plt.tight_layout()
        plt.show()

        plt.figure(figsize=(12,6))
        plt.plot(x1, y1, ".", ms=4)
        plt.title(fname.split("/")[-1])
        plt.xlabel("Az_Offset")
        plt.ylabel("RIGHT_POL")
        plt.tight_layout()
        plt.show()
```
